Notebook for visualizing results of PINN after training.

Define: 
- epochs: describes the total number of epochs the model was trained with.
- capture: describes after how many epochs data is captures for visualization.
- filename: file where the variables are stored afrer training of the model.
- filenam2: additional file where the variables are stored afrer training of the model. This file is only for plotting "plot_losses_comparisson" and/or "plot_lambdas_comparison"

Set "True" or "False":
- plot_comparison: Plots analytical and predicted solution contained in "filename".
- plot_derivatives: Plots derivatives of beam deflection from 1st-4th order contained in "filename".
- plot_losses: Plot physics and boundary condition losses along training contained in "filename".
- plot_lambdas: Plot value of lambdas (Scaling) contained in "filename".
- plot_losses_comparison: If filename2 is given, it plots the difference between losses along training of results contained in "filename" and "filename2".
- plot_lambdas_comparison: If filename2 is given, it plots the difference between lambda values along training of results contained in "filename" and "filename2".
- plot_difference: Plots the decay of the loss contained in filename.
